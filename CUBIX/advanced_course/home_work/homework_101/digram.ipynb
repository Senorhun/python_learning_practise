```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(np.array([3,7,6,12,16]))
plt.show()
```

```python
plt.plot(np.array([1,3,4,8,9]),np.array([3,7,6,12,16]))
plt.show()
```

```python
plt.plot(
    np.array([1,3,4,8,9]),
    np.array([3,7,6,12,16]),
    marker='o',
    markersize=20,
    markerfacecolor='red',
    linestyle='dashed',
    linewidth=1,
    color = 'green'
)
plt.show()
```

```python
plt.scatter(np.array([1,3,4,8,9]),np.array([3,7,6,12,16]))
plt.show()
```

```python
np.random.seed(197777)
plt.scatter(
    x = np.random.randint(100, size = 100),   # x 
    y = np.random.randint(100, size = 100),   # y
    c = np.random.randint(100, size = 100),   # color
    cmap = 'nipy_spectral',                   # color-map
    s = 10 * np.random.randint(100, size = (100)),  # size
    alpha = 0.5,                              # transparency
)
plt.colorbar()
plt.show()
```

```python
plt.bar(['alma','narancs','banán','meggy','citrom'],np.array([1,8,4,9,3]))
plt.show()
```

```python
plt.barh(['alma','narancs','banán','meggy','citrom'],np.array([1,8,4,9,3]))
plt.show()
```

```python
plt.pie(np.array([1,8,4,9,3]), labels=['alma','narancs','banán','meggy','citrom'])
plt.show()
```

```python
plt.pie(
    np.array([1,8,4,9,3]),
    labels=['alma','narancs','banán','meggy','citrom'],
    explode=[0.0,0.0,0.2,0.0,0.0]
)
plt.show()
```

```python
np.random.seed(197777)
plt.hist(np.random.normal(100,15,1000), bins=50)
plt.show()
```

```python
plt.plot(np.array([1,8,4,9,3]))
plt.plot(np.array([7,6,2,7,8]))
plt.show()
```

```python
gyümölcsök = ['alma','narancs','banán','meggy','citrom']
számok = np.array([1,8,4,9,3])

plt.subplot(2,2,1)
plt.plot(számok)

plt.subplot(2,2,2)
plt.scatter(gyümölcsök,számok)

plt.subplot(2,2,3)
plt.barh(gyümölcsök,számok)

plt.subplot(2,2,4)
plt.pie(számok, labels=gyümölcsök)

plt.show()
```

```python
plt.plot(np.array([1,3,4,8,9]), np.array([3,7,6,12,16]))
plt.title('Vonaldiagram', fontdict={'family':'serif','color':'blue','size':20})
plt.xlabel('x koordináta')
plt.ylabel('y koordináta')
plt.grid()
plt.show()
```

```python
gyümölcsök_df = pd.DataFrame({
    'név':['alma','narancs','banán','meggy','citrom'],
    'érték': [1,8,4,9,3],
})
gyümölcsök_df.plot(kind='barh', x='név', y = 'érték')
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
g = sns.pairplot(penguins, hue="species")
g.map_lower(sns.kdeplot)
plt.show()
```
